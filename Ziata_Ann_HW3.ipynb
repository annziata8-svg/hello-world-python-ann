# 1. Write program with while loop which finds the followings for the given values between 150 to 1000

```python
# 1a. How many odd numbers are between 150 and 1000 (include 150 and 1000)

start = 150
total = 0
while start <= 1000:
    if start % 2  > 0 :
        total = total + 1
        start = start + 1
    else :
        start = start + 1
print(total)
```

Every integer divided by two that is even will have 0 remainder, and odd will have remainder 1. for every odd number, I will add a count to the total. the loop goes until 1000.

```python
# 1b. Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)

start = 150
total = 0
while start <= 1000:
    if start % 2  > 0 :
        total = total + start
        start = start + 1
    else :
        start = start + 1
print(total)
```

Now, I add up all the start numbers to the total.

```python
# 1c.  Find the multiplication of odd numbers which are between 150 and 1000 (include 150 and 1000)

start = 150
total = 1
while start <= 1000:
    if start % 2  > 0 :
        total = total * start
        start = start + 1
    else :
        start = start + 1
print(total)
```

Now, I multiply the odd numbers. It's quite large!

```python
# 1d. Find the sum of square of odd numbers which are between 150 and 1000 (include 150 and 1000)

start = 150
total = 0
while start <= 1000:
    if start % 2  > 0 :
        total = total + start*start
        start = start + 1
    else :
        start = start + 1
print(total)
```

Same while loop, but adding together the squares of the odd numbers with the ** function.

# 2. Write program with for loop which finds the followings for the given values between 150 to 1000.

```python
# 2a. How many odd numbers are between 150 and 1000 (include 150 and 1000)

total = 0

for x in range (150, 1000+1) :
  if x % 2  > 0 :
    total = total + 1
print(total)
```

Using the % remainder method to find the odd numbers again, and using the for/ range loop function. I don't need to include 1000 because it's even, but did anyway. I made a count total every time a number is determined to be odd.

Same idea as before, using the for/range tool, adding x to the total each time to get the sum of everything.

```python
# 3.  Write a program so that for given M, it computes the sum s = MΣk=1 (1/k)

s = 0
m = int(input("Please enter a value "))
k = 1
while k <= m :
  s = s + 1/k
  k = k + 1
print(s)
```

The sum s starts at 0. M is the highest number in the range. It will do the final round when k=m.

```python
# 4. Write a program which repeatedly reads numbers until the user enters “done”.
# Once “done” is entered, print out the total, count, and average of the numbers.
# If the user enters anything other than a number, detect their mistake using try and except
# and print an error message and skip to the next number

total = 0
count = 0

while True:
  a = input("Please enter a number ")
  if a.lower() == "done" :
      break
  try  :
    total = total + float(a)
    count = count + 1
  except ValueError :
    print("Invalid input")

print(total, count, total/count)
```

The total and count start at 0. The while loop goes until "done". While running, the input will be added to the total. If the input is invalid (text), it will read as an error and will not effect the total or the loop.

```python
# 5. Write another program that prompts for a list of numbers as above (question 4)
# and at the end prints out both the maximum and minimum of the numbers instead of the average.

list = []

while True:
  a = input("Please enter a number ")
  if a.lower() == "done" :
      break
  try  :
    a = int(a)
    list.append(a)
  except ValueError :
    print("Invalid input")

print(max(list), min(list))
```
