 # 1. Write a program that considers given list and prints True if all of them are odd, and False otherwise.
# For example, if list L is given as L= [ 1, 5, 15], code will print True
# If list L is given as L= [ 2, 1, 5, 15], code will print False

```python
# 1.

L = input("Please enter a list of numbers separated with spaces ").split()

for x in L :
    x = int(x)
    if x%2 == 0 :
      print("False")
      break
else :
  print("True")
```

I used the split() function to make the input a list. I ran a loop that checks if the number is even. Once it hits an even number, it breaks.

# 2. Write a Python code which finds the smallest even and odd numbers in a given list.
(Use for loop and if conditions for this problem. Do not use existing codes. Use your own codes).
(First find the even and odd numbers and put them in different lists (use append function). Then
use sort function to find the smallest observation of each list.)
(First find the even and odd numbers and put them in different lists (use append function). Then
use sort function to find the smallest observation of each list.)

```python
# 2.

evens = []
odds = []
L = input("Please enter a list of numbers separated with spaces ").split()

for x in L :
    x = int(x)
    if x%2 == 0 :
      evens.append(x)
    if x%2 > 0 :
      odds.append(x)

evens.sort()
odds.sort()

print("The smallest odd number is ", odds[0])
print("The smallest even number is ", evens[0])
```

Again, asking for an input of numbers and saving as a list. Separating the odd and even numbers with .append. Then using the sort function to sort the list in ascending order. To print the first number on the list, use odds[0]

# 3. Write a Python code which finds common even values between given two lists A and B by using for loop and if statement?
If the lists are given as follows, list1=[10,66,45,13,18,8] list2=[8,17,5,66,18,22, 6,548], then the common even values are 8,66, and 18

```python
# 3.

list1 = [10, 66, 45, 13, 18, 8]
list2 = [8, 17, 5, 66, 18, 22, 6, 548]
list1evens = []
list2evens = []
common = []

for x in list1 :
    x = int(x)
    if x%2 == 0 :
      list1evens.append(x)
    if x%2 > 0 :
      continue

for x in list2 :
    x = int(x)
    if x%2 == 0 :
      list2evens.append(x)
    if x%2 > 0 :
      continue

for x in list1evens:
    if x in list2evens:
        common.append(x)

common.sort()
print(common)
```

First, using the append function to keep only the even numbers in each list. Then, using the append function to sort into the empty Common list while numbers were found in both, and printing the sorted list.

# 4. How to get the n th smallest even value of given array? (Use for loop for this problem. Do not use existing codes. Use your own codes).
For example, in given list [11,23,58,31,56,22,43,12,65,19], if n is defined as 3. The program will
print 56. (By using for loop and obtain the set of evens. By using another for loop, from the set
of evens remove (n-1) observations and break the loop and find the minimum observation of the
updated list)

11 23 58 31 56 22 43 12 65 19

```python
# 4.

L = input("Please enter a list of numbers separated with spaces ").split()
n = int(input("Please enter the nth value "))

Levens = []

for y in L :
    y = int(y)
    if y%2 == 0 :
      Levens.append(y)

for i in range(n-1):
    min_Levens = Levens[0]
    for x in Levens:
        if x < min_Levens:
            min_Levens = x
    Levens.remove(min_Levens)

    nth_smallest = Levens[0]
    for x in Levens:
        if x < nth_smallest:
            nth_smallest = x

print(nth_smallest)
```

First, want the user to input a list, and the nth value. Create an empty evens list to be filled, and append it with the even numbers.
Then, the code runs n-1 (2) times to remove the smallest value. It starts the first value [0]. If the next value is smaller, it updates it in min.Levens. The remove code removes smaller the value, in this case the smallest 2 (n-1).
The last loop searches for the smallest value in in the array and saves it as nth_smallest.
