```python
# 1. Compose a program that writes a standard Gaussian random variable. And calculate ð‘ value.

import random
import math
u = float(random.random())
v = float(random.random())
z = (math.sin(2*math.pi)*v)*(-2 * math.log(u))**.5
print("u = ", u)
print("v = ", v)
print("z = ", z)
```

Here, I had to import both random and math libraries. The math library was used for the math funcations sin and ln. I had it print all 3 variables just so we can see the numbers and how they change.

```python
# 2. Write a program that takes two integer values a and b from the command line and prints them in ascending order, separated by a space

a = int(input("Please enter one integer value "))
b = int(input("Please enter a second integer value "))
if a > b:
  print(b, a)
else :
  print(a, b)
```

```python
# 3. Write a program that takes three double values x, y, and z as command-line arguments
# and prints True if the values are strictly ascending or descending (i.e., x < y < z or x > y > z), and False otherwise

x = float(input("Please enter a double value: "))
y = float(input("Please enter a second double value: "))
z = float(input("Please enter a third double value: "))

if x < y < z or x > y > z:
  print("True")
else :
  print("False")
```

Here, the question wanted double values, which refers to float data types. The code will print true if the numbers are in ascending or descending order, using the OR function.

```python
# 4. Write a program that reads a magnitude from the user and displays the appropriate descriptor as part of a meaningful message.

mag = float(input("What is the magnitude of the earthquake? "))
if mag < 2.0:
  print("An earthquake with a magnitude of ", mag, "is considered a micro earthquake. ")
elif 2.0 <= mag < 3:
  print("An earthquake with a magnitude of ", mag, "is considered a very minor earthquake. ")
elif 3 <= mag < 4:
  print("An earthquake with a magnitude of ", mag, "is considered a minor earthquake. ")
elif 4 <= mag < 5:
  print("An earthquake with a magnitude of ", mag, "is considered a light earthquake. ")
elif 5 <= mag < 6:
  print("An earthquake with a magnitude of ", mag, "is considered a moderate earthquake. ")
elif 6 <= mag < 7:
  print("An earthquake with a magnitude of ", mag, "is considered a strong earthquake. ")
elif 7 <= mag < 8:
  print("An earthquake with a magnitude of ", mag, "is considered a major earthquake. ")
elif 8 <= mag < 10:
  print("An earthquake with a magnitude of ", mag, "is considered a great earthquake. ")
else :
  print("An earthquake with a magnitude of ", mag, "is considered a meteoric earthquake. ")
```

Here, ELIF is used for the many different options for the output. The magnitude input corresponded to different earthquake categories, printed as text strings. I would have also included code for negative earthquakes since the code will still print out "micro" although negative earthquakes, which can exist, are not perceptible by humans.
